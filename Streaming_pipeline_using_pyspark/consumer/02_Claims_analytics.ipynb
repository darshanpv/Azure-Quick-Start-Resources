## Use of Databricks Delta Tables and SQL for dashborads and visualization
### Sample analytical queries

Following quries will be used to create the dashboard
- What is a total number of claims served till date - (Number card tile)
- Counrty wise claims settlement - (GeoMap)
- What are the top 10 customers based on the claimed amount. (Table)
- What is breakup in terms of Hold, Rejected and Approved claims (Donut graph)

```
%sql
USE events_db;
SELECT * from claims_data;
```

```
%sql
SELECT count(*) FROM claims_data LIMIT 10;
```

```
%sql
SELECT country, count(*) AS total_claims
FROM claims_data
GROUP BY country
ORDER BY total_claims DESC
LIMIT 10;
```

```
%sql
SELECT DISTINCT customer_name,claim_amount
FROM claims_data
ORDER BY claim_amount DESC
LIMIT 10;
```

```
%sql
SELECT Status, count(*)AS Total
FROM claims_data
GROUP BY status;
```
